```python
%pip install langchain-openai dotenv
from langgraph.graph import StateGraph, START, END
import langchain_core
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
#state
class LLMstate(TypedDict):
    question : str
    ans : str
```

```python
def qa_LLM(state : LLMstate) -> LLMstate:
    
    #extract the question from the state
    question = state['question']
    #form a prompt
    prompt = f"Answer the following question {question}"
    
    #give the prompt to the LLM
    answer = model.invoke(prompt).content
    
    # update the enswer in the state
    state['ans'] = answer
    
    return state
```

```python
# create the graph
graph = StateGraph(LLMstate)

# create the node of the graph
graph.add_node('qa_LLM', qa_LLM)

# create the edged of the graph
graph.edges(START,'qa_LLM')
graph.edges('qa_LLM', END)

workflow = graph.compile()

graph.compile()
```

```python
#execute
initial_state = {'question' : 'Who is the best entrepreneur in the world'}

final_state = workflow.invoke(initial_state)

print(final_state)
```
